```python
query = """
SELECT *
FROM tweets
WHERE original_text LIKE '%vaccine%'
LIMIT 10;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    COUNT(CASE WHEN original_text LIKE '%vaccine%' THEN 1 END) AS vaccine,
    COUNT(CASE WHEN original_text LIKE '%lockdown%' THEN 1 END) AS lockdown,
    COUNT(CASE WHEN original_text LIKE '%government%' THEN 1 END) AS government
FROM tweets;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT sentiment, COUNT(*) AS count
FROM tweets
GROUP BY sentiment
ORDER BY count DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    'Vaccine' as topic, COUNT(*) as count
FROM tweets
WHERE clean_tweet LIKE '%vacc%'

UNION ALL

SELECT 
    'Lockdown', COUNT(*)
FROM tweets
WHERE clean_tweet LIKE '%lockdown%'

UNION ALL

SELECT 
    'Government', COUNT(*)
FROM tweets
WHERE clean_tweet LIKE '%government%';
"""
pd.read_sql(query, conn)
```

```python
### What can we deduce from the above:

#COVID-19 discourse in this dataset is dominated by discussions around lockdown measures,...
#followed by vaccination, while direct references to government are minimal.


#- Total: (2198 + 3214 + 15 = 5427)
#Vaccine (0): (2198/5427)*100 = 40.5%
#Lockdown (1): (3214/5427)*100 = 59.2%
#Government (2): (15/5427)*100 = 0.3%


#Lockdown-related discourse dominates COVID-19 conversations (40.5%), suggesting that restrictive measures were the primary focus of public discussion.
#Vaccine-related discourse is also substantial (59.2%), while initial keyword searches suggest limited direct reference to government (0.3%);
#however, expanding keyword definitions is necessary to fully capture political discourse, highlighting the importance of robust Boolean query design in social listening analysis.

#The use of the word/term "government" may not give exact estimates.The following query will aim to use as many related search terms as possible
```

```python
query = """
SELECT 
    'Government' as topic, COUNT(*) as count
FROM tweets
WHERE clean_tweet LIKE '%government%'
   OR clean_tweet LIKE '%president%'
   OR clean_tweet LIKE '%minister%'
   OR clean_tweet LIKE '%state%'
   OR clean_tweet LIKE '%authority%';
"""
pd.read_sql(query, conn)
```

```python
### What can we deduce from the above:

#Updated results: 0 (Vaccine) = 2198; 1 (Lockdown) = 3214; 2 (Government) = 5308


#- Total: (2198 + 3214 + 5308 = 10720)
#Vaccine (0): (2198/10720)*100 = 20.5%
#Lockdown (1): (3214/10720)*100 = 30.0%
#Government (2): (5308/10720)*100 = 49.5%

#Government-related discourse dominates COVID-19 conversations (49.5%), followed by lockdown (30.0%) and vaccination (20.5%) topics.
#This suggests that public sentiment is strongly centred on political leadership and institutional responses, rather than purely health-related themes.
#The large shift in results after expanding keywords highlights the importance of robust Boolean query design in accurately capturing social discourse.
```
